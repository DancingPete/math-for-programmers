# Imports

```python
# Standard library imports
from dataclasses import dataclass

# Third party imports
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Classes for Added Shorthand Functionality

```python
@dataclass
class Vector:
    x: int
    y: int

    def __add__(self, v):
        return Vector(self.x + v.x, self.y + v.y)

@dataclass
class Polygon:

    vectors: list[Vector]
    
    def __add__(self, v: Vector):
        return Polygon([Vector(vector.x + v.x, vector.y + v.y) for vector in self.vectors])

    def draw(self, ax: plt.Axes, color: str):

        vectors = self.vectors
    
        vectors.append(vectors[0])
        xx = [v.x for v in vectors]
        yy = [v.y for v in vectors]
    
        ax.set_aspect("equal")
        ax.set_xlim(-10, 10)
        ax.set_ylim(-10, 10)
        ax.set_xticks(range(-10, 11))
        ax.set_yticks(range(-10, 11))
        ax.grid(visible=True)
    
        for index in range(len(vectors) - 1):
    
            ax.plot(
                (vectors[index].x, vectors[index+1].x),
                (vectors[index].y, vectors[index+1].y),
                color=color,
            )
```

```python
fig, ax = plt.subplots()
vectors = list(map(lambda v: Vector(v[0], v[1]), [(6,4), (5,1), (1,-2), (2,-3), (1,-4), (-1,-4), (0,-3), (-1,0), (-2,1), (-4,0), (-5,1), (-2,2), (-5,2), (-5,3), (-4, 4), (-3, 4), (-2,5), (-1,5), (1,2), (3,1)]))
dino = Polygon(vectors)
dino.draw(ax, "red")
dino2 = dino + Vector(2, 2)
dino2.draw(ax, "blue")
```
